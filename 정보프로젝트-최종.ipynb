```python
import csv

f = open('시군구별 가해운전자 음주정도.csv')
data = csv.reader(f)
next(data)
next(data)
data = list(data)

result = []
result1 = []
for i in range(4,12) :
    result.append([])
    for row in data :
        if row[i] == '' :
            row[i] = 0
        result[i-4].append(int(row[i]))
    result1.append(sum(result[i-4])/len(result[i-4]))

print (result1)
    
def function(i):
    print (sum(result[i-4])/len(result[i-4]))
    
function(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8,3),dpi = 300)

plt.rc('font',family="Malgun Gothic")
plt.plot(result1)

plt.title('혈중 알코올 농도에 따른 사고발생횟수')

plt.xlim(0,7)
plt.ylim(0,60)

plt.xticks(range(8), ['0.05~0.09%','0.10~0.14%','0.15~0.19%','0.20~0.24%','0.25~0.29%','0.30~0.34%','0.35%이상','측정불응'])

plt.xlabel('음주농도')
plt.ylabel('사고발생횟수')

plt.savefig('data.png')

plt.show()
```
